```python
# Importing the Libraries 
# Importing MySql Connector Library 

import pandas as pd 
from sqlalchemy import create_engine
```

1. Data Loading

```python
# Loading CSV file 
df = pd.read_csv("Healthcare_Data.csv")
df.head(5)
```

2. Structure Check

```python

print( df.shape )       # displaying the No. of rows and columns 
print( df.columns )     # displaying the columns names 
print( df.dtypes )      # displaying data types 
print( df.info )        # displaying structure of dataset
```

3. Data Cleaning 

```python
print( df.isnull().sum() )  # Checking the null values in the dataset

# checking Duplicate Values 
print("Total Duplicates : ", df.duplicated().sum() )
print(len(df))

# Remove the Duplicates 
print(df[df.duplicated()])
df = df.drop_duplicates() 

print( "Total Duplicates : ",df.duplicated().sum(), "\nTotal Records = ", len(df) )
```

```python
# Renaming the Column Names
df = df.rename(columns = {'Name' : 'Patient_Name'}) 
df = df.rename(columns = {"Hospital" : "Hospital Name"}) 

# Convert the columns name in to Snake Case Format
df.columns = df.columns.str.replace(" ", "_")
df.head()
```

```python
#Reducing the length of colunm names 
df['Patient_Name'] = df['Patient_Name'].str.title()
df['Patient_Name'] = df['Patient_Name'].str.split().str[0]

df['Doctor'] = df['Doctor'].str.split().str[0]
df['Doctor'] = df['Doctor'].apply( lambda x:x if str(x).startswith('Dr') else 'Dr. ' + str(x) )
print(df['Patient_Name'], df['Doctor'])
```

```python
# Reduce the decimal numbers in Billing_Amount into two decimals 
df['Billing_Amount'] = df['Billing_Amount'].round(2) 
print(df['Billing_Amount'])
```

```python
# Counting the Total Patiences  
print("The Total Patients : ",df['Patient_Name'].count() )

# Counting Frequency of each unique values of Hospital_Name
print(df['Hospital_Name'].value_counts() )
```

```python
# Converting the "string based data columns" into "DataTime format" for time series opertions. 
print(df['Date_of_Admission']. dtype )

df['Date_of_Admission'] = pd.to_datetime( df['Date_of_Admission'] )   #converting 

df['Year'] = df['Date_of_Admission'].dt.year   # creating a new column to store the year 

df['Date_of_Admission'] = pd.to_datetime( df['Date_of_Admission'] ).dt.date

print(df['Year'])
```

```python
# Checking the size of values in Year 
print(df.groupby('Year').size())

# printing the count of values of blood_type, medical_condiion, and Name
print(df.groupby('Year')['Blood_Type'].value_counts())
print(df.groupby('Year')['Medical_Condition'].value_counts())
print(df.groupby('Year')['Patient_Name'].value_counts())
```

```python
# save the file after data cleaning 
df.to_csv("Cleaned_Healthcare_Data.csv", index = False)
```
